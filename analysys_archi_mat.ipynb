```python
import pandas as pd
from bokeh.plotting import figure
from bokeh.io import push_notebook,show, output_notebook
from bokeh.models import HoverTool
output_notebook()
```

```python

df_vect = pd.read_csv('./depot/result_vect.txt', sep=',',names=['Taille','Temps séquentiel','Temps parallèle','result/size']).iloc[:-200]
df_novect = pd.read_csv('./depot/result_no_vect.txt', sep=',',names=['Taille','Temps séquentiel','Temps parallèle','result/size']).iloc[:-200]

p = figure(title=f"Time to process", x_axis_label='Vector size', y_axis_label='Time (s)',width=1500, height=600,)
p.line(df_novect['Taille'], df_novect['Temps séquentiel'], legend_label="Temps sequentiel (no vect)", line_width=2,color='blue')
p.line(df_novect['Taille'], df_novect['Temps parallèle'], legend_label="Temps parallele (no vect)", line_width=2, color='yellow')
p.line(df_vect['Taille'], df_vect['Temps séquentiel'], legend_label="Temps sequentiel", line_width=2,color='red')
p.line(df_vect['Taille'], df_vect['Temps parallèle'], legend_label="Temps parallele", line_width=2, color='orange')
p.legend.location = "top_right"

show(p)
df_vect.head()
```

```python

df_vect = pd.read_csv('./depot/result_vect.txt', sep=',',names=['Taille','Temps séquentiel','Temps parallèle','result/size'])


p = figure(title=f"Time to process", x_axis_label='Vector size', y_axis_label='Time (s)',width=1500, height=600,)
p.line(df_vect['Taille'], df_vect['Temps séquentiel'], legend_label="Temps sequentiel", line_width=2,)
p.line(df_vect['Taille'], df_vect['Temps parallèle'], legend_label="Temps parallele", line_width=2, color='orange')
p.legend.location = "top_right"

show(p)
df_vect.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(25,8))
plt.plot(df_vect['Taille'], df_vect['Temps séquentiel'],label='Temps sequentiel')
plt.plot(df_vect['Taille'], df_vect['Temps parallèle'],legend='Temps parallele')
plt.legend('upper left')
plt.grid(True)
plt.xlabel('Vector size ')
plt.ylabel('Time (ns)')
plt.title('Time to process')
```
